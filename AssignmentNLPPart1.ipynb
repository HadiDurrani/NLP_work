```python
import nltk
import re
from pprint import pprint

string = 'AaBbCcDdEeFfGgHhIiJjKkLlMm!@#$%^&* 123.23 NnOoPpQqRrSsTtUuVvWwXxYyZzpat'

regex_patterns = [
    r'[a-zA-Z]+',  #Consecutive Upper or Lower case Characters
    r'[A-Z][a-z]*', #Single Upper case followed by 0 or more lower case characters
    r'p[aeiou]{,2}t', #Starting character p followed by 0 or less than 2 (0 <= aeiou >2) vowels and word ending in t
    r'\d+(\.\d+)?',    #A decimal character with . and floating points as optional
    r'([^aeiou][aeiou][^aeiou])*', #First character should be consonants and last should be consonant with a vowel enclosed between them this pattern can be repeated 0 or more times as denoted by *
    r'\w+|[^\w\s]+', #Tokenizes words and alphaneumeric characters ignoring non word and white space characters
]

for pattern in regex_patterns:
    print(pattern)
    nltk.re_show(pattern, string)
    print("\n")
```

Q3, Q4 and Q5

```python
import re
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem import PorterStemmer, LancasterStemmer
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup

def text_from_url_with_selenium(url):
    chrome_options = Options()
    chrome_options.add_argument('--headless')
    chrome_service = ChromeService(executable_path='C:\codes\cd\chromedriver.exe')
    driver = webdriver.Chrome(service=chrome_service, options=chrome_options)
    driver.get(url)
    driver.implicitly_wait(10)
    page_source = driver.page_source
    driver.quit()
    soup = BeautifulSoup(page_source, 'html.parser')
    text_without_html = soup.get_text()
    return text_without_html

url = 'https://www.csail.mit.edu/people/?roleFacets=Principal%20Investigators,Core%2FDual,Associates,Emeritus,Core%2FDual,Associates,Emeritus,Principal%20Investigators,Core%2FDual,Associates,Emeritus'
result = text_from_url_with_selenium(url)

if result is not None:
    print(result)

def find_phone_numbers_and_emails(text):
    phone_regex = r'\b(?:\d{3}[-.\s]?)?\d{3}[-.\s]?\d{4}\b'
    email_regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'

    phone_numbers = re.findall(phone_regex, text)
    emails = re.findall(email_regex, text)

    return phone_numbers, emails

def tokenize_text(text):
    tokenizer = RegexpTokenizer(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b|\b\w+\b')
    tokens = tokenizer.tokenize(text)
    return tokens

def stem_text(tokens):
    porter_stemmer = PorterStemmer()
    lancaster_stemmer = LancasterStemmer()
    porter_stems = [porter_stemmer.stem(token) for token in tokens]
    lancaster_stems = [lancaster_stemmer.stem(token) for token in tokens]

    return porter_stems, lancaster_stems

if result:
    tokens = tokenize_text(result)
    phone_numbers, emails = find_phone_numbers_and_emails(result)
    print("Phone numbers:")
    for phone_number in phone_numbers:
        print(phone_number)
    print("\nEmail addresses:")
    for email in emails:
        print(email)
    porter_stems, lancaster_stems = stem_text(tokens)
    print("\nPorter stems:")
    print(porter_stems)
    print("\nLancaster stems:")
    print(lancaster_stems)
```
